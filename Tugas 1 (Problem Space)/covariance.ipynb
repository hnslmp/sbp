```python
import pandas as pd
df = pd.read_excel (r'D:\Universitas Indonesia\Semester 6\Sistem Berbasis Pengetahuan (SBP)\Tugas 1\Dataset.xlsx')
print (df)
```

```python
from sklearn import preprocessing

score = df['Score'].tolist()
amin, amax = min(score), max(score)
for i, val in enumerate(score):
    score[i] = (val-amin) / (amax-amin)

gender = df['Gender'].tolist()
amin, amax = min(gender), max(gender)
for i, val in enumerate(gender):
    gender[i] = (val-amin) / (amax-amin)

ras = df['Ras'].tolist()
amin, amax = min(ras), max(ras)
for i, val in enumerate(ras):
    ras[i] = (val-amin) / (amax-amin)

parental = df['Parental'].tolist()
amin, amax = min(parental), max(parental)
for i, val in enumerate(parental):
    parental[i] = (val-amin) / (amax-amin)

lunch = df['Lunch'].tolist()
amin, amax = min(lunch), max(lunch)
for i, val in enumerate(lunch):
    lunch[i] = (val-amin) / (amax-amin)

tes = df['Test'].tolist()
amin, amax = min(tes), max(tes)
for i, val in enumerate(tes):
    tes[i] = (val-amin) / (amax-amin)

# np.ravel(score)
# np.ravel(gender)
# np.ravel(ras)
# np.ravel(parental)
# np.ravel(lunch)
# np.ravel(tes)

# score = preprocessing.normalize([score])
# gender = preprocessing.normalize([gender])
# ras = preprocessing.normalize([ras])
# parental = preprocessing.normalize([parental])
# lunch = preprocessing.normalize([lunch])
# tes = preprocessing.normalize([tes])

# score = score.tolist()
# gender = gender.tolist()
# ras = ras.tolist()
```
